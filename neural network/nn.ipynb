```python
from kerastuner import HyperModel
```

```python
class CNNHyperModel(HyperModel):
    def __init__(self, input_shape):
        self.input_shape = input_shape

    def build(self, hp):
        
        model = Sequential()
        model.add(Dense(units=hp.Int('unit', min_value=35, max_value=65, step=8, default=40),
                        activation=hp.Choice('dense_activation', values=['relu', 'tanh'], default='relu')))
        
        model.add(tf.keras.layers.BatchNormalization())
        
        model.add(Dense(units=hp.Int('lay_u', min_value=10, max_value=55, step=5, default=20),
                        activation=hp.Choice('dense_act', values=['relu', 'tanh'], default='tanh')))
        model.add(tf.keras.layers.BatchNormalization())
        
        model.add(Dense(1,activation='sigmoid',kernel_initializer='glorot_uniform',
            bias_initializer='zeros'))
        
        model.compile(
            optimizer=keras.optimizers.Adam(
                hp.Float(
                    'learning_rate',
                    min_value=1e-4,
                    max_value=1e-2,
                    sampling='LOG',
                    default=1e-3
                )
            ),
            loss='binary_crossentropy', metrics=['accuracy']
        )
        return model
```

```python
from kerastuner.tuners import Hyperband
from kerastuner.tuners import RandomSearch
```

```python
HYPERBAND_MAX_EPOCHS = 30
MAX_TRIALS = 10
EXECUTION_PER_TRIAL = 1
```

```python
hypermodel = CNNHyperModel(input_shape=x_train.shape)

tuner = RandomSearch(
    hypermodel,
    objective='val_accuracy',
    seed=1,
    max_trials=MAX_TRIALS,
    executions_per_trial=EXECUTION_PER_TRIAL,
    directory='random_search',
)
```

```python
tuner.search_space_summary()
```

```python
%%time
N_EPOCH_SEARCH = 30

tuner.search(x_train, y_train, epochs=N_EPOCH_SEARCH, validation_data = (x_test, y_test))
```

```python
# Show a summary of the search
tuner.results_summary()
```

```python
# Retrieve the best model.
best_model = tuner.get_best_models(num_models=1)[0]
```

```python
# Evaluate the best model.
best_model.evaluate(x_test, y_test, verbose=1)
```
